## Introduction
In this project, we model a swarm of UAV drones, tasked with mapping arbitrary targets while avoiding interspersed obstacles. On a high level the aim is to find an optimal set of parameters that maximises target mapping, minimises agents crashing into obstacles/each other and also minimises time taken. A detailed discussion of the cost function and parameters follows later in the report. Given the initial velocity and positions of the drones/targets/obstacles, we calculate the net force acting on each drone (henceforth referred to as agent), and use Forward Euler
time discretisation to repeatedly update its position and velocity. Genetic algorithms are then used to find optimal parameters



## Background and Theory

Given a fixed cartesian basis $e_1, e_2, e_3$, the position, velocity, and acceleration are respectively given by

$r = r_1e_1 + r_2e_2 + r_3e_3$

$v = \dot{r} = \dot{r_1}e_1 + \dot{r_2}e_2 + \dot{r_3}e_3$

$a = \ddot{r} = \ddot{r_1}e_1 + \ddot{r_2}e_2 + \ddot{r_3}e_3$

By Newton's Second Law, the net force acting on agent i is

$m_ia_i = \psi^{tot}_i \text{, where } \psi^{tot}_i = F_{p,i} + F_{d,i}$, denoting the propulsive and drag forces respectively

$F_{p,i} = F_{prop} * n^{*}_{i} \text{, where } F_{prop} = 200$ N is the magnitude of the propulsive force, and

$n^{*}_{i}$is the unit vector, which is found by adding 
all the weighted interaction vectors 
(agent-agent, agent-target, agent-obstacle),

and then normalised by dividing it by its Euclidean norm

$F_{d,i} = 1/2\rho_aC_{d,i}A_i\lVert (v_a - v_i) \rVert(v_a - v_i)$

Note: Please refer to appendix for glossary of symbols and constants

After calculating net force acting on an agent, we use Forward Euler time discretisation to find updated position and velocity

$v_i(t + \Delta t)  \dot{ = }  v_i(t) + a_i(t)\Delta t = v_i(t) + \psi^{tot}_i (t) \frac{\Delta t}{m_i}$

$r_i(t + \Delta t)  \dot{=}  r_i(t) + v_i(t)\Delta t$

To find terminal velocity v_t we set $\psi^{tot}_i = 0$  and solve, using the values provided in the project spec. Rearranging terms, we get  

$v_t = \sqrt{\frac{2F_{prop}}{\rho_aC_{d,i}A_i}} = 36.14 \text{ m/s}$


As noted above, the key equation here is the use of Forward Euler time discretisation, which allows us to solve for updated positions and velocities of each agent based on net acceleration and previous values of r and v. This is a simple numerical scheme that is easy to implement. However, since this is an explicit method where everything happens at the same time (ie. an agent at a certain time t0 cannot use information from any other agent at the time t1), we must use small time step t. Using a larger value of t has the benefit that each agent will be able to take larger steps towards targets/away from objects as needed, but the disadvantage is that agents may crash into one another due to these larger steps, which could have been avoided by using smaller t instead. The inability to safely use large t for Forward Euler time discretisation is its primary drawback

It is worth noting that these equations are based on a point-mass idealisation of the agents. For this to be a reasonable assumption to make, the agents themselves should not be too large in terms of volume, and each agent's mass should be at least uniformly distributed, or preferably concentrated near its center already. The main consequence of point mass idealisation is that we ignore the effect of rotation on the agent's center of mass in our model. On the topic of the model, we note that the only two forces it currently accounts for are the propulsion and drag forces. As such, if we were to introduce a bias to our model ie. a force that is pointed in a specific direction that isn't already accounted for in our 2 forces, then our model would not be as well suited for optimisation. Hence, we would expect higher costs if we introduced additional forces in the form of net gravity/lift/buoyancy

## Procedure and Methods:
A brief note on the cost function before returning to discussing the model: The cost function used here is
$\Pi = w_1M^* +w_2T^* + w_3L^*$
There are multiple reasons why gradient-based methods (numerical or otherwise) are not suitable to optimise this cost function: the main reason is that the cost function itself is highly non-convex, and additionally, there's too much variation between different simulations (especially since drone and target positions are randomised), and attempting to analytically or symbolically differentiate across 15 parameters is needlessly computationally expensive. Genetic algorithms are suitable because we have a better chance of escaping local non global minima, and since we're including the parent strings, we know cost will stay constant or decrease amongst parents from generation to generation. Additionally, we're not necessarily interested in reaching the absolute global minima provided we end up in the right region or a region of similarly low cost, which makes genetic algorithms suitable again. 

Returning to the model, we note that our control model is a simplification of the real world. More complex models may be able to give results that are lower cost and or more suited to real world constraints, but this does not mean that the model used here is not physically accurate given the assumptions made. As such, we expect our model to work for paramteter values where the situation being modelled is closely in line with the simplifications and assumptions made here. For example, in cases where the drones are very small and spherical in shape in an environment where there are winds producing a lift that negates the effect of gravity, we would expect our model to work for some parameter values
On the topic of model parameters, if any of the a, b, c parameters became negative, then we would expect our model to perform poorly, since instead of having attractive and repulsive interaction terms that exponentially decay with distance, the term where any of a, b, c became negative would increase exponentially with distance, meaning we would give agents/obstacles/targets that are far away greater importance, which we should avoid. To "remove" crashed agents/mapped targets, I set their position vectors to Inf whenever they crash/are mapped. Then, whenever I am iterating through a new agent, I check if the target/other agent had been set to Inf, and simply continue on to the next iteration if that is the case. If the agent itself has crashed, then I break the relevant loop instead of continuing. To avoid the issue of agents crashing into each other at the beginning, I initialised their positions as far apart from each other as possible while still respecting the region provided. Please check appendix for code. Before moving on to the next section of the report, I would like to make mention of 2 modifications in my code: To minimise inbreeding in my genetic algorithm, I introduced mutations. Specifically, all my random  $\Phi_{i}$ were between -0.5 and 1.5

Additionally, I ran into the issue of my agents spiralling around targets instead of actually approaching them and collecting them To fix this, I multiplied my drag force by a factor of 5

## Conclusion
The goal of this project was to use genetic algorithms to find a set of parameters to govern the motion of a swarm of UAV drones with a primary emphasis on mapping targets and a secondary emphashis on minimising time and # of agents crashed. The key equation to this model was Forward Euler time discretisation, which allowed us to update velocity and position vectors of agents based on initial values and the sum of all forces acting on them. After modelling the system (with a few simplifications such as point mass idealisation and neglecting forces such as gravity), we were able to come up with a set of paramters with an associated cost of ~10% which was able to map all the targets in allotted time. While the model and genetic algorithm could probably be improved, we note here that even this simplified model was effective at helping us achieve our goal with relatively low cost. An unexpected result was the neagative parameter in Design 3, wherein there were no repulsive agent-agent forces. While may one assume this would result in excessive agent-agent collisions, we see that this was actually one of the best performing strings overall
