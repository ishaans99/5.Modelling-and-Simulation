Before modelling the system, a few clarifications need to be made. As shown in the figure, the y-axis is actually the vertical axis here; all references to the co-ordinate system will be based on the conventions depicted in the image. Individual rods are named L1, L2, and L3. As is visible in the schematic (but reiterated here for clarity), L1 and L2 are in the x-y plane, while L3 is in the x-x plane. In our setup, all 3 rods have constant angular velocity $$\dot{\theta}$$
where $\theta$ is angular position at time t, and initial angular position is given by $\theta_0$. This does not mean that the angular velocities of all arms are equal, but rather that the angular velocity for a given arm is unchanging throghout the simulation. Bearing this in mind, we can now derive expressions for coordinates in terms of defined variables 

Note: We use subscripts to specify which arm we're referring to.


As stated above, $\theta$ is a function of time. Specifically, $$\theta_j = \theta_j(t) = \theta_j^0 + \dot{\theta}_jt, j = \{1,2,3\}$$

We note that since L1 is in the x-z plane with an angular position of $\theta_1$, its component in the x-direction is the cosine of this angle, multiplied by the length of 

L1 itself. This stems naturally from the definition of cosine. As such, $x_d = L_1\cos{\theta_1}$. However, since we have multiple rods to account for, the overall equations (by analogous argument to the one made above), become:

$x_d = L_1\cos{\theta_1} + L_2\cos{\theta_2} + L_3\sin{\theta_3}$

$y_d = L_1\sin{\theta_1} + L_2\sin{\theta_2}$

$z_d = L_3\sin{\theta_3}$

To obtain an expression for velocities, we simply diffferentiate these positions with respect to time, remebering that $\dot{\theta}$ is angular velocity. Then

$\dot{x}_d = -L_1 \dot{\theta}_1 \sin{\theta_1} - L_2 \dot{\theta}_2 \sin{\theta_2} + L_3 \dot{\theta}_3 \cos{\theta_3}$

$\dot{y}_d = L_1 \dot{\theta}_1 \cos{\theta_1} + L_2 \dot{\theta}_2 \cos{\theta_2}$

$\dot{z}_d = - L_3 \dot{\theta}_3 \sin{\theta_3}$

Our velocity vector $v_d$ comprises these 3 components

Analogous to the above discussion on angular position and velocity, let initial droplet position and velocity be denoted as 

$r_i^0$ and $v_i^0$, with position and velocity at that time denoted as $r^d$ and $v^d$. As such, our equations become
$r_i^0 = r^d$ and $v_i^0 = v^d + \Delta v^d$, where $\Delta v^d$ accounts for the relative droplet dispenser velocity that helps propel the droplet out of the dispenser.

As in past projects, we use Forward Euler Integration to iteratively update our velocities and positions. The equations are as follows:

$r_i(t + \Delta t) = r_i(t) + \Delta tv_i(t)$

$v_i(t + \Delta t) = v_i(t) + \Delta t\ddot{r}_i(t)$

We have initial values of position and velocity to start with, but to be able to perform Forward Euler integration we must calculate acceleration $\ddot{r}_i(t)$ for a given droplet at a given time. To do so, we find the net force for given droplet. We model each deposited droplet as a lumped mass that moves as a solid sphere of radius R. As such, the total forces equation becomes 

$m_i\ddot{r}_i = \psi^{tot}_i = F^{grav}_i + F^{elec}_i + F^{drag}_i$

Here $m_i$ denotes mass of the droplet, and right hand side of the equation refers to gravitational, electrostatic, and drag forces, respectively.

Before we calculate mass, we need to clarify some assumptions and make intermediate calculations. Specifically, we assume that the droplets are a mixture of two different phases ie. two different materials. The volume fraction of the second material is given by $v_2$, and the individual densities and charge capacities of the 2 phases are denoted by $\rho_1, \rho_2, q_1, q_2$, respectively. 

With these, we calculate effective density and electrical charge capacity per unit volume, $p^*, q^*$, respectively using the equations:

$p^* = (1-v_2)\rho_1 + v_2\rho_2$

$q^* = (1-v_2)\langle q_1 \rangle + v_2 \langle q_2 \rangle$

We assume that every droplet is spherical with radius R = 0.001 m, and calculate volume $V_i = \frac{4 \pi R^3}{3}$ Using this, we are able to calculate lumped droplet charge and mass

$q_i = V_iq^*$

$m_i = V_i\rho^*$

Since we are assuming all droplets are identical in this regard, we calculate these values only once, and use them for any given droplet as and when needed. 

Returning to the total forces equation, we now focus on each individual term:
Calculating $F^{grav}_i$ is fairly straightforward. In our co-ordinate system, the y term pertains to vertical orientation, so it will be the only nonzero term here. 
$F^{grav}_i = [0; -m_ig; 0]$, where g = 9.81

In order to calculate net electrical force, we recall that droplets are modelled as lump charges with charge $q_i.$ Additionally, the aforementioned charged substrate is modelled as a set of $N_c = 100$ fixed-position point charges
with charge and position $q_p = -8*10^{-5}, r_p$, respectively. (Check appendix for code used to generate grid). Hence,
$$F^{elec}_i = \sum_{p=1}^{N_c} \frac{q_pq_i (r_i - r_p)}{4 \pi \epsilon \|r_i - r_p \|^3}, \epsilon =\epsilon_0 (constant)$$


Note: This equation differs from the one in the project spec. I'm pretty sure the vectors in $(r_i - r_p)$ 

needed to be normalised, and as such, I divided the expression by an additional (meaning not in the project spec) factor of $\|r_i - r_p \|$ to do so


In order to calculate drag force, we need information about the geometry of the droplets and of the surrounding medium. Since we're modelling the droplets as spheres, drag reference area and velocity $A^D_i = \pi r^2, v_i$ respectively

Regarding the surrounding medium, we denote important information using the following variables:
$\rho_a =$ density
$C_{Di} =$ drag coefficient 
$v^{ f} =$ velocity
$\mu_f =$ viscosity 

With these terms defined, we calculate drag force as:

$$F^{drag}_i = \frac{1}{2} \rho_a C_{Di} \|v^{ f} - v_i \|^2 (v^{ f} - v_i)  A^D_i$$

Note: We determine our value of the drag coefficient, we must determine, for each droplet, the corresponding Reynolds number $Re$

$$Re = \frac{2R\rho_a \|v^{ f} - v_i \|}{\mu_f}$$

Check appendix to see how drag coefficient is calculated based on $Re$

In terms of implementing this in code, we follow the necessary steps to determine initial position and velocity. After this, we carry out the calculations needed to calculate total force acting on each droplet, and consequently use Forward Euler Intergration to update our positions and velocity. We repeat this iteratively till all the drops land on the printing bed.

The gravitational force acts only in the vertical direction. However, both the drag and electrostatic forces act in all directions on each droplet. If each droplet had only a gravitational force acting on it, it should fall straight down, and each point should be directly below where the extruder was when the droplet was released. Due to the eletric grid points, which are equally scattered throughout the bed, the droplets are attracted towards them and move in the direction where net attraction is strongest. Similarly, the drag force affects how long the droplet is in the air, with a stronger drag force resulting in longer time periods for the electrostatic force to act on the drople before it hits the printer bed. The symmetric nature of the eletric grid points is responsible for why the droplet pattern is more spread out than the extruder pattern, while still retaining the original shape

To examine the effect of the electrostatic force, I created rdiff = riwith-riwithout, where the former is a collection of the ri vectors with electrostatic force, while the latter is a collection without it. By observing the colour bar, we note that initially, the electrostatic force is pulling droplets to the lower left quadrant of the grid, while later it is pushing them to the uppder right quadrant. By correlating this information with Figure 2, we see that the electrostatic force is always pulling droplets to the center, thereby helping ensure the droplets aren't deposited off the surface of the printer bed. Depending on the specific angles of the extruder and the medium, drag force may hinder a droplets path to a desired location on the printer bed, and the electrostatic force helps counter this

## Genetic Algorithm

The model we feed into the genetic algorithm is simplified. We set charges = 0 and assume drag force is negligible, thereby leaving us with only the gravitational force to consider. Given the simplified nature of this model, we can arrive at an analytical solution fairly easily, without relying on Forward Euler Integration.

By the equations of motion:
$S(t) = S(0) ut + \frac{1}{2}at^2$
S(t): Displacement at time = t 
S(0): Initial position
u: initial velocity
a = acceleration
t = time
We want the t where the droplet hits printer bed, where vertical distance is given by $r_{i2}$. Hence, we solve for optimal time $t^*$
$S(t) = 0, a= -g, u = v_{i2}$ 
Solving, we get
$t^* = (-v_{i2} - (v_{i2}^2 + 2*g* r_{i2}^0))/-g$
We ignore negative solution since time can only be positive
After computing time, we simply update using $r_i = r_i^0 + v_i^0*t^*$
Note that second component = 0 for all vectors

For cost, we compute arc length by comparing the total norm between the position vectors in a current generartion and the ideal vectors provided to us, normalised by the total norm between successive ideal vectors
